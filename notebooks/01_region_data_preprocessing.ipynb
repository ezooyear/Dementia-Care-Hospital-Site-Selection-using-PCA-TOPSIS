```python
# 제외할 '시군구' 값 목록
exclude_values = [
    '전국', '서울특별시', '부산광역시', '인천광역시', '대구광역시', '광주광역시', '대전광역시', '울산광역시',
    '세종특별자치시', '경기도', '충청북도', '충청남도', '전라남도', '전라북도', '경상북도',
    '경상남도', '강원도', '제주특별자치도'
]

# '시군구' 값이 exclude_values에 포함되지 않은 행만 남김
df_main = df_main[~df_main['시군구'].isin(exclude_values)]

# '시도' 열에서 '전국' 값을 가진 행을 삭제
df_main = df_main[df_main['시도'] != '전국']

# '시도' 값에 따라 '시군구' 값에 추가할 접두어 정의
prefixes = {
    '서울특별시': '서울_',
    '부산광역시': '부산_',
    '대구광역시': '대구_',
    '인천광역시': '인천_',
    '광주광역시': '광주_',
    '대전광역시': '대전_',
    '울산광역시': '울산_',
    '경기도': '경기_',
    '강원도': '강원_',
    '충청북도': '충북_',
    '충청남도': '충남_',
    '전라북도': '전북_',
    '전라남도': '전남_',
    '경상북도': '경북_',
    '경상남도': '경남_',
    '세종특별자치시' : '세종_',
    '제주특별자치도': '제주_'
}

# '시도' 값에 따라 '시군구' 값에 접두어 추가
df_main['시군구'] = df_main.apply(lambda row: prefixes[row['시도']] + row['시군구'], axis=1)
```

### B. 독거노인 데이터 전처리

```python
# 선택된 두 개의 열만 포함하는 데이터프레임 생성
selected_columns = ['행정구역', '독거노인가구비율']
final_df_alone = df_alone[selected_columns]

new_column_names = {
    '행정구역': '시군구',
    '독거노인가구비율': '독거노인가구비율'
}

# 열 이름 변경
final_df_alone = final_df_alone.rename(columns=new_column_names)
```

```python
# 시도와 접두어 매핑
prefixes = {
    '서울특별시': '서울_',
    '부산광역시': '부산_',
    '대구광역시': '대구_',
    '인천광역시': '인천_',
    '광주광역시': '광주_',
    '대전광역시': '대전_',
    '울산광역시': '울산_',
    '경기도': '경기_',
    '강원도': '강원_',
    '충청북도': '충북_',
    '충청남도': '충남_',
    '전라북도': '전북_',
    '전라남도': '전남_',
    '경상북도': '경북_',
    '경상남도': '경남_',
    '세종특별자치시' : '세종_',
    '제주특별자치도': '제주_'
}

# 접두어를 추가하는 함수
def add_prefix(row, current_prefix):
    if row['시군구'] in prefixes:
        current_prefix[0] = prefixes[row['시군구']]
        return row['시군구']
    elif current_prefix[0] != '':
        return current_prefix[0] + row['시군구']
    else:
        return row['시군구']

# 초기 접두어
current_prefix = ['']

# 데이터프레임의 '시군구' 열 업데이트
final_df_alone['시군구'] = final_df_alone.apply(add_prefix, axis=1, current_prefix=current_prefix)

# 수정된 데이터프레임 출력
final_df_alone
```
